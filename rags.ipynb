```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI, OpenAIEmbeddings


```

```python

pdfs = PyMuPDFLoader('unsus.pdf')
docs = pdfs.load()
docs
```

### pdf 다음은? 문서 분할

```python
text_splits = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 50)
split_docs = text_splits.split_documents(docs)
```

### 1000글자 넘는 건 없었나보다

```python
text_splits = RecursiveCharacterTextSplitter(chunk_size = 400, chunk_overlap = 40)
split_docs = text_splits.split_documents(docs)
```

```python
len(split_docs)
```

### 문단을 쪼개서 10->27개로 쪼갰음

```python
print(split_docs[1].page_content)
```

### 이제 쪼갠 것을 임베딩해야함

```python
embeddings = OpenAIEmbeddings()
embeddings 
```

### 임베딩 다음엔? 벡터로 저장하기 이지만 딱히 저장은 안한다는 거 커널죽으면 사라질예정

```python
vectordb= FAISS.from_documents(documents = split_docs, embedding = embeddings)
```

### faiss라는 벡터 db에서 적합한 문서를 들고와줄 리트라이버를 선언

```python
retriever = vectordb.as_retriever()
retriever
```

### 얘도 인보크로 호출할 수 있음, 결과를 보면 llm의 답변이 아닌 그냥 문서를 들고옴, 이걸 의미 단위(dense)로할지, keyword단위로 할 것 인지에 따라 어떠한 리트라이버? 리트리버? 리트리버라고 하자 리트리버를 쓸지 달라짐

```python
retriever.invoke('아내가 원하던 음식이 무엇인가요?')
```

### 이제 리트리버를 써서 템플릿을 쓸 거니까 템플릿의 초안을 작성해보자

```python
prompt = PromptTemplate.from_template(
    ''' you're an assistant for question-answering task.
    use the following pieces of retrieved context to answer the question 
    if you don't know the answer, just say that you don't know.
    answer is Korean.
    #Question:
    {question}
    
    #context:
    {context}
    
    #answer:'''
)

llm = ChatOpenAI(model='gpt-4.1-nano-2025-04-14', temperature=0)
```

### 체인을 묶어보자 

```python
chain = {'context':retriever,'question':RunnablePassthrough()} | prompt | llm 
```

### 인보크에 뭐 어디 변수에 뭐 어떻게 넣을 건지 딕셔너리 형태로 명시 안했지? 병렬로 두 변수에 넣어짐

```python
chain.invoke('아내가 가장 먹고 싶어하던 음식과, 왜 그 음식을 먹지 못했는지 알려줘')
```

### 프롬프트에 몇가지 추가해보자

```python
prompt = PromptTemplate.from_template(
    ''' you're an assistant for question-answering task.
    use the following pieces of retrieved context to answer the question 
    if you don't know the answer, just say that you don't know.
    you must include 'producer' name and 'page' number in your answer
    answer is Korean.
    #Question:
    {question}
    
    #context:
    {context}
    
    #answer:'''
)
```

```python
chain = {'context':retriever,'question':RunnablePassthrough()} | prompt | llm 
chain.invoke('아내가 무엇을 먹고 싶어했으며, 왜 먹지 못했는지 알려줘')
```
